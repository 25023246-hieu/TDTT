```python
#w7a1
def binary_search(n,x) :
    r  = len(n)-1 
    l = 0
    while l <= r :
        m = int ((r+ l)/2)
        if n[m] == x :
            return m 
        elif n[m]  < x :
            l = m +1
        elif n[m] > x :
            r = m -1
    return -1
while True :
    a = list( map (int , input().split()))
    k = int (input())
    print(binary_search(a,k))

        
```

```python
#w7a2
def count_occurrences(a,k) :
    dem = 0
    for i in a :
        if i == k :
            dem += 1 
    return dem 
while True :
    a = list(map (int, input().split()))
    k =int (input())
    print(count_occurrences(a,k))
```

```python
#w7a3
def bubble_sort( a) :
    for i in range ( len(a)) :
        for j in range( i , len(a)-1) :
            if a[j] > a[j+1] :
                a[j] , a[j+1] = a[j+1] , a[j] 
    return a 
def selection_sort(a) :
    for i in range (len(a)) :
        min = i 
        for j in range ( i , len(a)) :
            if a[min] > a[j] :
                min = j
        a[i] , a[min] = a[min] , a[i]
    return a
a = list(map (int , input ().split()))
print( bubble_sort(a))
print( selection_sort(a))
```

```python
#w7a4
def search(a) :
    dict_a = {}
    max=0
    ans=0
    for i in  a :
        if i not in dict_a :
            dict_a[i] = 1
        elif i in dict_a :
            dict_a[i] += 1
    for i in dict_a :
        if dict_a[i] > max :
            max = dict_a[i]
            ans = i
        if dict_a[i] == max :
            if i < ans :
                ans = i 
    return ans,"xuat hien nhieu nhat",max,"lan"
a = list( map (int , input().split()))
print(search(a))
```

```python
#w7a5
def sum(a , k) :
    dem = 0
    for i in  range (len(a)) :
        for j in range (i, len(a)) :
            if a[i] + a[j] == k :
                dem += 1 
    return dem 
a = list (map (int , input().split()))
k = int(input())
print (sum (a,k))
```
