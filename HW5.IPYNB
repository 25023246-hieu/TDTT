```python
#w6a21
def find_str(s) :
    dict={}
    for i in  range (len(s)):
        if s[i] not in dict :
            dict[s[i]] =[i] 
        else :
            dict[s[i]].append(i)
    for i in dict :
        dict[i] = tuple(dict[i])
    return dict
s = list( map (str, input().split()))
print(find_str(s))
```

```python
#w6a24
def first_zero( n) :
    a=[]
    dem = 0
    for i in n :
        if i !=0 :
            a.append(i)
        else :
            dem += 1 
    for i in range (dem) :
        a.insert(0,0)
    return a
def end_zero(n):
    m = first_zero(n)
    m.reverse()
    return m

n =  list (map(int ,input(). split()))

print(first_zero(n))
print(end_zero(n))
```

```python
#w6a26
def sum_maxi(n) :
    sum = 0
    if len(n) < 5 :
        return 0
    else :
        for i in range (2 , len(n)-2) :
            if n[i] > n[i-1] and n[i] > n[i+1] :
                sum += n[i]
    return sum
n = int (input())
a = list( map (int , input().split()))
print (sum_maxi(a))


```
